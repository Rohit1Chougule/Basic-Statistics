# **Q11)  Suppose we want to estimate the average weight of an adult male in    Mexico. We draw a random sample of 2,000 men from a population of 3,000,000 men and weigh them. We find that the average person in our sample weighs 200 pounds, and the standard deviation of the sample is 30 pounds. Calculate 94%,98%,96% confidence interval?**

```python
from scipy.stats import norm

# Define confidence levels
confidence_levels = [0.94, 0.96, 0.98]

# Calculate critical values
critical_values = {}
for level in confidence_levels:
    critical_values[level] = norm.ppf((1 + level) / 2)

print("Critical values for different confidence levels:")
for level, value in critical_values.items():
    print(f"Confidence level: {level}, Critical value: {value:.3f}")
```

```python
import numpy as np

# Given data
sample_mean = 200  # Sample mean
sample_std = 30  # Sample standard deviation
sample_size = 2000  # Sample size

# Z-scores for different confidence levels
z_scores = {
    94: 1.881,
    98: 2.326,
    96: 2.054
}

# Calculate confidence intervals
for confidence_level, z_score in z_scores.items():
    margin_of_error = z_score * (sample_std / np.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    print(f"{confidence_level}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

## ***Q 22) Calculate the Z scores of  90% confidence interval,94% confidence interval, 60% confidence interval ***

```python
from scipy.stats import norm

# Confidence levels
confidence_levels = [0.90, 0.94, 0.60]

# Calculate Z-scores for each confidence level
for confidence_level in confidence_levels:
    z_score = norm.ppf((1 + confidence_level) / 2)
    print(f"{confidence_level*100}% Confidence Interval Z-score:", z_score)
```

## **Q 23) Calculate the t scores of 95% confidence interval, 96% confidence interval, 99% confidence interval for sample size of 25**

```python
from scipy.stats import t

# Sample size
n = 25

# Degrees of freedom
df = n - 1

# Confidence levels
confidence_intervals = [95, 96, 99]

# Calculate t-scores for each confidence level
for confidence_level in confidence_intervals:
    # Find the t-score for the given confidence level and degrees of freedom
    t_score = t.ppf((1 + confidence_level/100) / 2,df)
    print(f"{confidence_level}% confidence interval t-score :", t_score)
```
